Following Dr. Arpin's Example from Lecture 1
===

The example from Lecture 1 was to consider $p=103$, $E: y^2 = x^3-x$ over $\overline{\mathbb{F}_p}$.

We have that $\mathrm{End}(E) \subseteq B_{p,\infty} = \mathbb{Q} + \mathbb{Q}i + \mathbb{Q}j + \mathbb{Q}ij$ with $i^2 = -1$, $j^2 = -p$, and $ij = -ji$.  Consider $\mathrm{End}(E) = \mathbb{Z} + \mathbb{Z}i + \mathbb{Z} \frac{1+j}{2} + \mathbb{Z}\frac{i+ ij}{2}$, with $j: (x,y)\mapsto(x^p, y^p)$ and $i: (x,y) \mapsto (-x, \sqrt{-1}y)$.

My note/not from lecture: $-1$ should be taken as "the map that performs multiplication by $-1$ in the elliptic curve" (and similar for $-p$).  This is why $i^2 = (x, -y)$ --- this is the negation map on an elliptic curve.

```python
p = 103
Q.<i,j,k> = QuaternionAlgebra(-1, -p)
```

Dr. Arpin claimed that $\mathrm{End}(E)$ corresponded with a maximal order in $B_{p,\infty}$, and I want to justify this to myself via computation. (There is a theorem that shows this in general for $\mathrm{End}(E)$.)

From Dr. Eisenträeger's lecture *today*, we now know how to determine that an order is maximal: namely, we consider the discriminant of the order:
$$\mathrm{disc}(\mathcal{O}) := \left| \det\right(\mathrm{Trd}(\alpha_i\alpha_j )_{i,j} \left) \right|,$$
where the reduced trace $\mathrm{Trd}(\alpha_i\alpha_j) = \alpha + \overline{\alpha}$.

(This is mentioned on page 7 of the notes, just after Definition 2.1.2; also look at Voight, section 14.5: *Ramification and discriminant*).

Finally, we know that the order is maximal if and only if the reduced discriminant (i.e. the square root of the discriminant) is $p$.

So, we need to find this discriminant for our order $O$.

```python
basis = [1, i, (1+j)/2, (i+i*j)/2]
```

```python
M = Matrix(QQ, 4, 4, lambda i, j: basis[i]*basis[j] + (basis[i]*basis[j]).conjugate())
pretty_print(M)
```

```python
disc_O = sqrt(abs(M.det()))
disc_O
```

Thus, we can see that the order that's isomorphic to $\mathrm{End}(E)$ for our specific curve is maximal, which agrees with the claim from lecture that $\mathrm{End}(E)$ is always maximal.

To do this a bit less "by hand," we can use the `quaternion_order` function in Sage:

```python
O = Q.quaternion_order(basis)
print(O.discriminant())
print(f"Is O maximal? {'yes' if O.is_maximal() else no}")
```
